```python
import cv2
import matplotlib.pyplot as plt

img_path = r"D:\SmartVision\DataSets\smartvision_dataset\detection\images\image_000010.jpg"
label_path = r"D:\SmartVision\DataSets\smartvision_dataset\detection\labels\image_000010.txt"

img = cv2.imread(img_path)

if img is None:
    print("❌ Image not found → check path")
else:
    h, w, _ = img.shape

    with open(label_path) as f:
        for line in f:
            cls, x, y, bw, bh = map(float, line.split())

            x1 = int((x - bw/2) * w)
            y1 = int((y - bh/2) * h)
            x2 = int((x + bw/2) * w)
            y2 = int((y + bh/2) * h)

            cv2.rectangle(img, (x1, y1), (x2, y2), (0,255,0), 2)

    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis("off")
    plt.show()
```
