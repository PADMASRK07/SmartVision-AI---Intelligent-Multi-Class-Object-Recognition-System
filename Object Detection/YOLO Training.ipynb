```python
# ==============================================================================
# STEP 3.3: YOLOv8 TRAINING & FINE-TUNING
# ==============================================================================
# This cell installs the Ultralytics framework, verifies GPU acceleration, 
# configures the hyperparameter landscape, and executes the training loop.
# ==============================================================================

# 1. Install and Import Ultralytics Framework
%pip install -q ultralytics

import os
import torch
from ultralytics import YOLO

# 2. Hardware Verification
device = "0" if torch.cuda.is_available() else "cpu"
print(f"🚀 Execution Device Detected: {device.upper()}")
if device == "0":
    print(f"🔥 GPU Model: {torch.cuda.get_device_name(0)}")
else:
    print("⚠️ WARNING: CPU detected. Training YOLOv8 on CPU will be extremely slow!")

# 3. Path Configuration
# Ensure this points to the exact path where your data.yaml file resides
DATA_YAML_PATH = os.path.abspath(r"D:\SmartVision\DataSets\smartvision_dataset\detection\data.yaml")
    

if not os.path.exists(DATA_YAML_PATH):
    raise FileNotFoundError(f"❌ Missing 'data.yaml' at: {DATA_YAML_PATH}. Please check Phase 3.2 data setup.")

# 4. Initialize Pre-trained YOLOv8 Model
# We use 'yolov8n.pt' (YOLOv8 Nano) to guarantee high real-time inference speeds (30-50 FPS)
# while utilizing pre-trained COCO feature-extraction weights.
print("\n📥 Loading pre-trained YOLOv8 Nano model...")
model = YOLO("yolov8n.pt")

# 5. Define Training Parameters & Hyperparameters
TRAIN_PARAMS = {
    "data": DATA_YAML_PATH,       # Path to dataset configuration yaml
    "epochs": 50,                 # Recommended standard for stable convergence on a 25-class subset
    "batch": 16,                  # Balanced batch size to avoid GPU Out-Of-Memory (OOM) errors
    "imgsz": 640,                 # Standard native evaluation resolution for YOLOv8
    "optimizer": "AdamW",         # AdamW offers robust weight decay handling for transfer learning
    "lr0": 0.01,                  # Initial learning rate
    "lrf": 0.01,                  # Final learning rate fraction (cosine decay)
    "device": device,             # GPU id '0' or 'cpu'
    "workers": 4,                 # Number of worker threads for data loading
    "project": "SmartVision_AI",  # Project folder name
    "name": "yolov8_25_classes",  # Run experiment folder name
    "save": True,                 # Automatically checkpoint and save best weights
    "plots": True,                # Generates loss curves, precision-recall metrics, and confusion matrix
    "val": True                   # Run verification on the validation dataset at every epoch end
}

# 6. Execute Fine-Tuning Run
print("\n🏁 Starting YOLOv8 Training Loop...")
print("-" * 60)

try:
    results = model.train(**TRAIN_PARAMS)
    print("-" * 60)
    print("✅ Training successfully completed!")
    
    # 7. Model Asset Locations & Output Summary
    # Weights are automatically saved in: {project}/{name}/weights/
    best_weights_path = os.path.join("SmartVision_AI", "yolov8_25_classes", "weights", "best.pt")
    print(f"\n🏆 Best Target Model Weights Saved At: {os.path.abspath(best_weights_path)}")
    print("📈 Metric visualizations and performance charts are generated in the 'SmartVision_AI/yolov8_25_classes/' directory.")

except Exception as e:
    print(f"\n❌ An error occurred during training: {str(e)}")
```

```python
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO

# 1. Load the model instance explicitly
model = YOLO("yolov8n.pt")

# 2. Force confidence threshold to 1% to print raw candidate features
img_path = r"D:\SmartVision\DataSets\smartvision_dataset\detection\images\image_001982.jpg"

results = model.predict(
    source=img_path,
    conf=0.01,    # Drop threshold to 1% to see what the model is ignoring
    imgsz=640,
    save=False
)

# 3. Inspect raw prediction structures
boxes = results[0].boxes
print("\n🔍 === RAW MODEL CANDIDATE INSPECTION ===")
print(f"Total raw detection boxes found above 1% confidence: {len(boxes)}")

if len(boxes) > 0:
    for idx, box in enumerate(boxes):
        cls_id = int(box.cls[0])
        conf_score = float(box.conf[0])
        # Look up standard COCO class name
        model_label = model.names[cls_id]
        print(f"👉 Candidate {idx+1}: Class ID {cls_id} ({model_label}) | Confidence: {conf_score:.2%}")
else:
    print("❌ Absolute zero detections found. The model sees no recognizable COCO features here.")
```
